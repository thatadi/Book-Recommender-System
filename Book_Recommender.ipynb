```python
ratings_title = ratings.merge(books,on='ISBN')
ratings_title
```

```python
ratings_title.info()
```

```python
ratings_title.isnull().sum()
```

```python
x=ratings_title.groupby('User-ID').count()['Book-Rating'] >= 200
```

```python
critics=x[x].index
```

```python
y=ratings_title[ratings_title['User-ID'].isin(critics)]
```

```python
new=y.groupby('ISBN').count()['Book-Rating'] >=50
famous_books=new[new].index
```

```python
filtered=y[y['ISBN'].isin(famous_books)]
```

```python
final=filtered.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
final.fillna(0,inplace=True)
final
```

```python
sim_scores=cosine_similarity(final)
sim_scores.shape
```

```python
def recom(book_name):
  index=np.where(final.index==book_name)[0][0]
  distances=sim_scores[index]
  z=list(enumerate(distances))
  z=sorted(z,key=lambda x:x[1],reverse=True)[1:4]
  for i in z:
    print(final.index[i[0]])
```

```python
recom('Harry Potter and the Chamber of Secrets (Book 2)')
```
